```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("dark")
```

```python
all_data_file = "Data\Inputs_Pesticides_Trade_E_All_Data_(Normalized)\Inputs_Pesticides_Trade_E_All_Data_(Normalized).csv"
areacodes_file = "Data\Inputs_Pesticides_Trade_E_All_Data_(Normalized)\Inputs_Pesticides_Trade_E_AreaCodes.csv"
flags_file = "Data\Inputs_Pesticides_Trade_E_All_Data_(Normalized)\Inputs_Pesticides_Trade_E_Flags.csv"
itemcodes_file = "Data\Inputs_Pesticides_Trade_E_All_Data_(Normalized)\Inputs_Pesticides_Trade_E_ItemCodes.csv"
```

```python
data = pd.read_csv(all_data_file, encoding = "ISO-8859-1")
```

```python
eu_countries = ["Austria", "Belgium", "Bulgaria", "Croatia", "Cyprus", "Czech Republic", "Denmark", "Estonia", "Finland", "France", "Germany", "Greece", "Hungary", "Ireland", "Italy", "Latvia", "Lithuania", "Luxembourg", "Malta", "Netherlands", "Poland", "Portugal", "Romania", "Slovakia", "Slovenia", "Spain", "Sweden"]
eu_data = data[data.Area.isin(eu_countries)]
```

```python
eu_data_import = eu_data[eu_data.Element == "Import Quantity"]
eu_data_import = eu_data_import[eu_data_import.Item.isin(["Insecticides", "Herbicides", "Fungicides"])]
```

```python
eu_data_import[eu_data_import["Area"] == "Denmark"].sort_values("Year")
```

```python
# plotting all countries in one plot
for country in eu_countries:
    sns.lineplot(eu_data_import[eu_data_import["Area"] == country], x = "Year", y = "Value", hue = "Item")
```

```python
def lineplot(dataset, element, item):
    # element should be a string with one element
    # item should be a list
    # countries should be filtered first
    try:
        type(item) == list
    except:
        print("item is not a list")
    dataset = dataset[dataset.Element == str(element)]
    dataset = dataset[dataset.Item.isin(item)]
    g = sns.FacetGrid(dataset, col = "Area", hue = "Item", height = 4, aspect = 1.2, col_wrap= 2, sharex = False)
    g.fig.suptitle(str(element) + " of " + str(item), fontsize=16, y = 1)
    g.map(sns.lineplot, "Year", "Value")
    g.set_xticklabels(rotation=45)
    for ax in g.axes.flat:
        ax.set_xlim(1990, 2006)
        ax.legend(loc = "upper left")
```

```python
pesticides = ["Insecticides", "Herbicides", "Fungicides"]
lineplot(eu_data, "Export Value", pesticides)
```
